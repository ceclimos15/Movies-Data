```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('IMDB-Movie-Data.csv') 
data
```

```python
#first 10 rows of the dataset
data.head(10)
```

```python
#last 10 rows of the dataset
data.tail(10)
```

```python
#shape of the dataset
print('nb of rows:',data.shape[0])
print('nb of columns:',data.shape[1])
```

```python
#get information about our dataset
data.info()
```

```python
#check null values in the dataset
x = data.isnull()
sns.heatmap(x)
plt.show()
```

```python
#drop missing values
data = data.dropna(axis=0)
```

```python
#check for duplicate data
duplicate = data.duplicated().any()
duplicate
```

```python
#get the statistics about the data 
data.describe()
```

```python
# Display Title of The Movie Having Runtime >= 180 Minutes

x = data[data['Runtime (Minutes)']>= 180]
new = x['Title']
print('The movies with running time >= 180 are:')
new
```

```python
# In Which Year There Was The Highest Voting? Draw a bar graph showing this.

max_votes = data['Votes'].max()
highest_vote = data[data['Votes'] == max_votes]
year_highest_vote = highest_vote['Year'].iloc[0]
year_highest_vote

plt.figure(figsize=(10,6))
plt.bar(data['Year'],data['Votes'],color = 'red')
plt.bar(year_highest_vote, max_votes, color='blue', label=f'Highest Vote: {max_votes}')

plt.xlabel('Year')
plt.ylabel('Votes per year')
plt.title('Votes per Year with Highlighted Highest Vote')
plt.legend()

plt.show()
```

```python
#In Which Year There Was The Highest Revenue?

highest_revenue = data['Revenue (Millions)'].max()
highest_revenue_year = data[data['Revenue (Millions)'] == highest_revenue]['Year'].iloc[0]
highest_revenue_year

plt.bar(data['Year'] , data['Revenue (Millions)'], color = 'blue')
plt.bar(highest_revenue_year , highest_revenue , color = 'red' ,label = f'Highest revenue :{highest_revenue_year}')
plt.xlabel('Year')
plt.ylabel('Revenue (Millions)')
plt.legend()
plt.show()
```

```python
#  Find The Average Rating For Each Director
avg_director = data.groupby('Director')['Rating'].mean().sort_values()
print('The avg rating for each director are:')
avg_director
```
